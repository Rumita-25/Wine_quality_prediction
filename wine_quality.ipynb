```python
# deviding the data into dependant and independant variables

target='quality'

#target feature
y=white_data[target]

#input feature
x=white_data.drop(target, axis=1)
```

```python
# Scalling the data

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_res= sc.fit_transform(x)
```

```python
x1= x.drop(['residual sugar','density'],  axis=1)
x1.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(x1)
rescale_x= scaler.transform(x1)
```

```python
rescale_x.shape
```

```python
y.value_counts()
```

```python
# creating TRAIN & TEST  datasets

from sklearn.model_selection import train_test_split
# splitting x and y into training and testing

x_train, x_test, y_train, y_test= train_test_split(rescale_x, y, test_size= 0.1, random_state=8)  # 10% for the test data & 90% for train data
```

```python
model_res= pd.DataFrame(columns=['Model', 'Score'])
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
# Model  Building
#--------------------
# Decision tree classifier

from sklearn.tree import DecisionTreeClassifier
d_tree= DecisionTreeClassifier()
d_tree.fit(x_train, y_train)
d_train_pred= d_tree.predict(x_train)
d_test_pred= d_tree.predict(x_test)
```

```python
# Building Confusion Matrix and printing train and test accuracy
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print("TRAIN ACCURACY", accuracy_score(y_train, d_train_pred)*100)
print("TEST ACCURACY", accuracy_score(y_test, d_test_pred)*100)
```

```python
# Confusion matrix based on wine quality (3,4,5,6,7,8,9)  # Diagonal elements are correctly predicted
print(confusion_matrix(y_test, d_test_pred))
```

```python
# Classification Report
print(classification_report(y_test, d_test_pred, digits=5))
```

```python
# Macro average used to the system's performance across the data sets

from sklearn.metrics import precision_score, recall_score, classification_report, f1_score
print("PRECISION SCORE OF MACRO IS", round(precision_score(y_test, d_test_pred, average='macro')*100,3))
print("PRECISION SCORE OF MICRO IS", round(precision_score(y_test, d_test_pred, average='micro')*100,3))
print("PRECISION SCORE OF WEIGHTED IS", round(precision_score(y_test, d_test_pred, average='weighted')*100,3))
```

```python
# RECALL
print("RECALL SCORE OF MACRO IS: ", round(recall_score(y_test, d_test_pred, average='macro')*100,3))
print("RECALL SCORE OF MiCRO IS: ", round(recall_score(y_test, d_test_pred, average='micro')*100,3))
print("RECALL SCORE OF WEIGHTED IS: ", round(recall_score(y_test, d_test_pred, average='weighted')*100,3))
```

```python
# F1 SCORE
print("F1 SCORE OF MACRO IS: ", round(f1_score(y_test, d_test_pred, average='macro')*100,3))
print("F1 SCORE OF MICRO IS: ", round(f1_score(y_test, d_test_pred, average='micro')*100,3))
print("F1 SCORE OF WEIGHTED IS: ", round(f1_score(y_test, d_test_pred, average='weighted')*100,3))
```
